### Problems Identified within the Data

- 541 duplicated Records where found in the Data.

- Insufficient information were recorded for the house_rules and Licence columns.

- Dollar signs and commas recorded along with the values for price and service fee.

- The currency for price and service fee were not added to the column's titles.

- Missing values in some records

- Columns like the price, service fee, id, host id, last review, Construction year are assigned wrong datatypes.

- In the neighbourhood group column the spelling of 'Brooklyn' was misspelt in some records.

- Some outliers were identified in the availability 365 column.

### CLEAN THE DATA

Steps to be Taken for Data Cleaning
- Drop duplicated records.

- Drop house_rules and license columns with insufficient data.

- R emove all dollar signs in the price and service fee columns.

- Remove all commas from the price and service fee columns.

- Rename the price and service fee columns to include a dollar signs.

- Drop all records with missing values.

- Change all mismatched data types to the appropriate ones.

- Correct the spelling of 'brookln' to 'Brooklyn'.

- Get rid of outliners in the 'availability 365' column data.

```python
# Drop the duplicate records
df.drop_duplicates(inplace=True)

# Drop house_rules and license columns with insufficient data (ignore if not present)
df.drop(['house_rules', 'license'], axis=1, inplace=True, errors='ignore')

# Rename the price and service fee columns to include a dollar sign
df.rename(columns={'price': 'price ($)', 'service fee': 'service_fee ($)'}, inplace=True)

# Remove all dollar signs and commas in the price and service fee columns (only if dtype is object)
for col in ['price ($)', 'service_fee ($)']:
	if df[col].dtype == 'object':
		df[col] = df[col].str.replace('$', '', regex=False)
		df[col] = df[col].str.replace(',', '', regex=False)

# Drop all records with missing values
df.dropna(inplace=True)

# Change all mismatched datatypes to the appropriate ones
df['price ($)'] = df['price ($)'].astype(float)
df['service_fee ($)'] = df['service_fee ($)'].astype(float)
df['id'] = df['id'].astype(str)
df['host id'] = df['host id'].astype(str)
df['last review'] = pd.to_datetime(df['last review'])
df['Construction year'] = df['Construction year'].astype(int)

# Correct the spelling of 'brookln' to 'Brooklyn'
df.loc[df['neighbourhood group'] == 'brookln', 'neighbourhood group'] = 'Brooklyn'

# Get rid of outliers in the 'availability 365' column data
df = df.drop(df[df['availability 365'] > 500].index)

```

### What are the different property types in the Datasets?

```python
# Room type count plotted on a barchart 

room_type_bar = plt.bar(property_types.index, property_types.loc[:, "count"] );
plt.bar_label(room_type_bar, labels = property_types.loc[:, "count"], padding= 4 );
plt.ylim([0, 50000]);
plt.xlabel("Room Type");
plt.ylabel("Room Type Count");
plt.title("Property Types and their Counts in the Dataset");
```

- Private room

- Hotel room

Notably, the barchart illustrates that Entire home/apt dominates the dataset with 44,163 occurences, followed by Private room with 37,494 occurrences. Shared room and Hotel room trail behind with 1,646 and 108 occurrences, respectively.

##### Which neighborhoods group have the highest average prices for Airbnb listings?

```python
# Which neighborhood group have the highest average prices for Airbnb listings?
avg_price = df.groupby('neighbourhood group')['price ($)'].mean().sort_values(ascending=False).to_frame()
avg_price_bar = plt.bar(avg_price.index, avg_price.loc[:, "price ($)"] );
plt.bar_label(avg_price_bar, labels = round(avg_price.loc[:, "price ($)"], 2), label_type='edge', padding= 4 );
plt.ylim([0, 700]);
plt.xlabel("Neighborhood Groups");
plt.ylabel("Average Price per listings ($)");
plt.xticks(rotation=45);
plt.title("Average Price per Listings in each Neighborhood Groups ($)");
```

Among the five identified neighborhood groups in the dataset, Queens boasts the highest average listings average listings price at 630.46 USD, closely trailed by the Bronx with an average listings price of 630.20 USD. Conversely, Staten island records the lowest average listings price at 633.13 USD.

#### Are hosts with verified identies more likely to receive positive reviews?

```python
base_color = sns.color_palette()[0]
sns.boxenplot(data = df, x = 'host_identity_verified', y = 'review rate number', color = base_color);
plt.xlabel("Host Verification Status");
plt.ylabel("Review Rate Number");
plt.title("Average Review Rate for each Verification Status");
```

The average review rate for verified and unconfirmed hosts are 3.27 and 3.28, respectively, indicating a marginal difference in their review rates, with unconfirmed hosts having a slightly higher rate. This suggests that verified hosts are not necessarily more likely to receive position review compared to unconfirmed hosts.

#### Is there a correlation between the price of a listings and its servicce fee?

```python
# Is there a correlation between the price of a listings and its servicce fee?

df['price ($)'].corr(df['service_fee ($)'])
```
